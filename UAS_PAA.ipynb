```python
import time

def bubble_sort(arr):
    n = len(arr)
    start_time = time.time()
    for i in range(n):
        for j in range(n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
        if i < 5 or i >= n-5:
            print(f"Iterasi {i+1}: {arr}")
    end_time = time.time()
    return arr, end_time - start_time

def insertion_sort(arr):
    n = len(arr)
    start_time = time.time()
    for i in range(1, n):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j+1] = arr[j]
            j -= 1
        arr[j+1] = key
        if i < 5 or i >= n-5:
            print(f"Iterasi {i+1}: {arr}")
    end_time = time.time()
    return arr, end_time - start_time

def analyze_algorithm(arr):
    print("Sebelum pengurutan:", arr)
    print("")

    choice = None
    while choice != 3:
        print("Menu:")
        print("1. Bubble Sort")
        print("2. Insertion Sort")
        print("3. Keluar")
        choice = int(input("Pilih algoritma pengurutan atau masukkan 3 untuk keluar: "))

        if choice == 1:
            print("Bubble Sort:")
            sorted_arr, bubble_time = bubble_sort(arr.copy())
            print(f"\nSetelah pengurutan: {sorted_arr}")
            print(f"Waktu komputasi: {bubble_time:.8f} detik")
        elif choice == 2:
            print("Insertion Sort:")
            sorted_arr, insertion_time = insertion_sort(arr.copy())
            print(f"\nSetelah pengurutan: {sorted_arr}")
            print(f"Waktu komputasi: {insertion_time:.8f} detik")
        elif choice == 3:
            print("Keluar")
        else:
            print("Pilihan tidak valid. Silakan pilih 1, 2, atau 3.")

        print("")
# Main Program
list_data = [12, 99, 62, 15, 20, 95, 39, 48, 3, 24, 8, 43, 74, 19, 97, 33, 49, 68, 55, 33, 90, 90, 7,
             26, 85, 46, 39, 40, 9, 36, 60, 64, 89,31, 25, 71, 21, 23, 63, 84, 32, 5, 3, 44, 21, 10, 21,
             17, 50, 2, 36, 53, 79, 54, 19, 88, 1, 32, 31, 15, 6, 3, 1, 40, 22, 43, 18, 1, 77, 9, 59, 40, 7, 41, 81]
analyze_algorithm(list_data)
```

```python
import sys
import time

class Graph:
    def __init__(self, vertices):
        self.V = vertices
        self.graph = [[0 for _ in range(vertices)] for _ in range(vertices)]

    def add_edge(self, src, dest, weight):
        self.graph[src][dest] = weight
        self.graph[dest][src] = weight

    def tsp(self):
        start = 0
        visited = [False] * self.V
        path = [0] * (self.V + 1)
        path[0] = start
        visited[start] = True
        min_dist = sys.maxsize
        start_time = time.time()
        self.tsp_util(start, visited, path, 1, 0, min_dist)
        end_time = time.time()
        print(f"\nComputation Time: {end_time - start_time:.6f} seconds")

    def tsp_util(self, curr, visited, path, count, dist, min_dist):
        if count == self.V:
            if self.graph[curr][0] != 0:
                curr_dist = dist + self.graph[curr][0]
                if curr_dist < min_dist:
                    min_dist = curr_dist
                    self.print_path(path, min_dist)
            return

        for v in range(self.V):
            if not visited[v] and self.graph[curr][v] != 0:
                visited[v] = True
                path[count] = v
                self.tsp_util(v, visited, path, count + 1, dist + self.graph[curr][v], min_dist)
                visited[v] = False

    def print_path(self, path, dist):
        print("Shortest Path:", end=" ")
        for i in range(self.V):
            print(chr(ord('a') + path[i]), end=" ")
        print(chr(ord('a') + path[0]), end=" ")
        print(f"\nDistance: {dist}")

    def dijkstra(self):
        start = 0
        dist = [sys.maxsize] * self.V
        dist[start] = 0
        visited = [False] * self.V
        start_time = time.time()

        for _ in range(self.V):
            u = self.min_distance(dist, visited)
            visited[u] = True
            self.print_dijkstra(dist)
            print("")
            for v in range(self.V):
                if not visited[v] and self.graph[u][v] != 0 and dist[u] + self.graph[u][v] < dist[v]:
                    dist[v] = dist[u] + self.graph[u][v]
                    self.print_dijkstra(dist)

        end_time = time.time()
        self.print_dijkstra(dist)
        print(f"\nComputation Time: {end_time - start_time:.6f} seconds")

    def min_distance(self, dist, visited):
        min_dist = sys.maxsize
        min_index = -1
        for v in range(self.V):
            if not visited[v] and dist[v] < min_dist:
                min_dist = dist[v]
                min_index = v
        return min_index

    def print_dijkstra(self, dist):
        print("Shortest Distance from 'a' to:")
        for v in range(self.V):
            print(f"{chr(ord('a') + v)}: {dist[v]}")

def main():
    g = Graph(7)
    g.add_edge(0, 1, 12)
    g.add_edge(0, 2, 10)
    g.add_edge(0, 6, 12)
    g.add_edge(1, 3, 12)
    g.add_edge(1, 2, 8)
    g.add_edge(2, 3, 11)
    g.add_edge(2, 4, 3)
    g.add_edge(2, 6, 9)
    g.add_edge(3, 4, 11)
    g.add_edge(3, 5, 10)
    g.add_edge(4, 5, 6)
    g.add_edge(4, 6, 7)
    g.add_edge(5, 6, 9)

    print("Graph:")
    for i in range(g.V):
        for j in range(g.V):
            if g.graph[i][j] != 0:
                print(f"{chr(ord('a') + i)} to {chr(ord('a') + j)}: {g.graph[i][j]}")

    print("\nMenu:")
    print("1. Travelling Salesman Problem (TSP)")
    print("2. Dijkstra's Algorithm")
    choice = int(input("Enter your choice (1 or 2): "))

    if choice == 1:
        print("\n--- TSP ---")
        g.tsp()
    elif choice == 2:
        print("\n--- Dijkstra's Algorithm ---")
        g.dijkstra()
    else:
        print("Invalid choice. Exiting the program.")

if __name__ == "__main__":
    main()
```
